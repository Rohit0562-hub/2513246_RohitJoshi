```python
#Problem 4

#1
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

I = np.eye(2, dtype=int)

A_inv = np.linalg.inv(A)
AdotA_inv = np.dot(A, A_inv)

AdotA_inv_int = np.round(AdotA_inv).astype(int)
print(AdotA_inv_int)

if np.array_equal(AdotA_inv_int, I):
    print("True")
else:
    print(False)

#2
AdotB = np.dot(A,B)
BdotA = np.dot(B,A)

print(AdotB)
print(BdotA)

if np.array_equal(AdotB, BdotA):
    print("Equal")
else:
    print("Not Equal")


#3
AdotB = np.dot(A,B)
AdotB_T = AdotB.T

B_T = B.T
A_T = A.T
B_TdotA_T = np.dot(B_T, A_T)

if np.array_equal(AdotB_T, B_TdotA_T):
    print("Equal: (AB)^T = B^T A^T")
else:
    print("NotEqual")
    
```

```python
#Element-Wise Addition

import time
import numpy as np

n = 1_00_000

list1 = list(range(n))
list2 = list(range(n))

start_time = time.time()
list_sum = [list1[i] + list2[i] for i in range(n)]
end_time = time.time()
print("Time taken for Python list addition:", end_time - start_time, "seconds")

arr1 = np.array(n)
arr2 = np.array(n)

start_time = time.time()
arr_sum = arr1 + arr2
end_time = time.time()
print("Time taken for NumPy array addition:", end_time - start_time, "seconds")
```

```python
#Element wise Multiplication

import time
import numpy as np

n = 1_00_000

list1 = list(range(n))
list2 = list(range(n))

start_time = time.time()
list_product = [list1[i] * list2[i] for i in range(n)]
end_time = time.time()
print("Time taken for Python list multiplication:", end_time - start_time, "seconds")

arr1 = np.arange(n)
arr2 = np.arange(n)

start_time = time.time()
arr_product = arr1 * arr2
end_time = time.time()
print("Time taken for NumPy array multiplication:", end_time - start_time, "seconds")
```

```python
#Dot Product

import time
import numpy as np

n = 1_00_000

list1 = list(range(n))
list2 = list(range(n))

start_time = time.time()
dot_list = sum(list1[i] * list2[i] for i in range(n))
end_time = time.time()
print("Time taken for Python list dot product:", end_time - start_time, "seconds")

arr1 = np.arange(n)
arr2 = np.arange(n)

start_time = time.time()
dot_arr = np.dot(arr1, arr2)
end_time = time.time()
print("Time taken for NumPy array dot product:", end_time - start_time, "seconds")
```

```python
import time
import numpy as np
import random

n = 1000

matrix1 = [[random.randint(0,10) for i in range(n)] for i in range(n)]
matrix2 = [[random.randint(0,10) for i in range(n)] for i in range(n)]

start_time = time.time()

result_list = [[0]*n for i in range(n)]

for i in range(n):
    for j in range(n):
        for k in range(n):
            result_list[i][j] += matrix1[i][k] + matrix2[k][j]

end_time = time.time()
print("Time taken for Python list Matrix Multiplication", end_time - start_time , "seconds")

arr1 = np.random.randint(0, 10, size=(n,n))
arr2 = np.random.randint(0, 10, size=(n,n))

start_time = time.time()
result_np = np.dot(arr1, arr2)
end_time = time.time()
print("Time taken for Numpy Array Matrix Multiplication", end_time - start_time , "seconds")
```
